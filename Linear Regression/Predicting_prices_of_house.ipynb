```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
%matplotlib inline
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Linear Regression/house.csv")
X=df["sqft_living"]
Y=df["price"]
```

```python
df.plot(kind="scatter", x="sqft_living", y="price", color="red")
plt.show()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2)
```

```python
plt.scatter(X_train, Y_train, label="Training Data", color="red")
plt.scatter(X_test, Y_test, label="Testing Data", color="blue")
plt.legend()
plt.show()
```

```python
LR=LinearRegression()
```

```python
LR.fit(X_test.values.reshape(-1,1), Y_test.values)
```

```python
prediction=LR.predict(X_test.values.reshape(-1,1))
```

```python
plt.plot(X_test, prediction, label="Linear Regression", color="blue")
plt.scatter(X_test, Y_test, label="Actual Test Data", color="red", alpha=0.7)
plt.legend()
plt.show()
```
